# **Count how many people have the first name John**

# **Find the surname of everyone whose first name is more than 6 characters long**

```python
data[['FirstName', 'LastName']] = data['EmployeeName'].str.split(' ', n=1, expand=True) # Changed from 3 positional arguments to 2 and changed expand=True to a keyword argument.
```

```python
data
```

```python
# Filter to find last names of people whose first name is longer than 6 characters
long_first_names = data[data['FirstName'].str.len() > 6]['LastName']
long_first_names
```

# **Create a new derived column called overtime_ratio in the DataFrame which is the proportion of overtime pay to base pay**

```python
#Create the derived column 'overtime_ratio' as the proportion of OvertimePay to BasePay
data['overtime_ratio'] = data['OvertimePay'] / data['BasePay']
data
```

# **In one line, create a new DataFrame called new_df which contains only people who earn more than 100K base salary**

```python
new_df = data[data['BasePay'] > 100000]
new_df
```

# **Now create a new derived column in new_df that is the sum of their pays, not including base salary**

```python
# Create the derived column 'total_other_pays' as the sum of OvertimePay, Bonus, and OtherPay
new_df['total_other_pays'] = new_df[['OvertimePay', 'OtherPay']].sum(axis=1)
new_df
```

## Append a new row to the DataFrame to represent a made up employee (you can use the pd.Concat() method for this)
He/she has no benefits, so set this attribute to not a number using np.nan
# **Make up the other column values, but make sure they are standardised to the same convention as the other values in that column**

```python
# New employee data
new_employee = pd.DataFrame({
    'EmployeeName': ['ALEX NEWMAN'],
    'BasePay': [95000],
    'OvertimePay': [0],
    'OtherPay': [0]
})
import numpy as np # import the numpy module using the common alias np
# Set benefits columns to NaN
new_employee['OvertimePay'] = np.nan
new_employee['OtherPay'] = np.nan

# Append the new employee to the original DataFrame
data_updated = pd.concat([data, new_employee], ignore_index=True)
data_updated
```
